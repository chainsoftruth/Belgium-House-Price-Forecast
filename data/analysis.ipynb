What is the correlation between the variables and the price? (Why might that be?)

legend :
- -1  → Negative relation
- 0   → No relation
- 1   → Strong relation

```python
corr_liza = liza.corr(numeric_only=True)

corr_liza
```

How are variables correlated to each other? (Why?)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(corr_liza*100, annot=True, cmap="coolwarm")
plt.title("Correlation in % for each aspect of the properties")

plt.show()
```

legend :
- -1  → Negative relation
- 0   → No relation
- 1   → Strong relation

Which variables have the greatest influence on the price?
