```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
def _subtype_replace(str):
    match str:
        case "studio" | "chalet":
            return 1
        case "ground-floor" | "apartment" | "bungalow" | "triplex" | "duplex" | "residence":
            return 2
        case "mixed-building" | "loft" | "cottage" | "penthouse" | "master-house" | "villa":
            return 3
        case "mansion":
            return 4
```

```python
df = pd.read_csv("data/clean_dataset.csv")
# df.type = df.type.apply(lambda x: 1 if x == "house" else 2)
# df["subtype"] = df["subtype"].apply(lambda x: _subtype_replace(x))
```

```python
_houses = df[df.type == 1].iloc[:,1:]
_apartments = df[df.type == 2].iloc[:,1:]
_apartments = _apartments.drop("land_area", axis = 1)
```

```python
hy = _houses.price
apy = _apartments.price
hX = _houses.drop("price", axis=1)
apX = _apartments.drop("price", axis=1)
hX_train, hX_test, hy_train, hy_test = train_test_split(hX, hy, random_state = 24, test_size = 0.25)
apX_train, apX_test, apy_train, apy_test = train_test_split(apX, apy, random_state = 24, test_size = 0.25)
```

```python
h_scaler = StandardScaler()
h_scaler.fit(hX_train)
hX_train = h_scaler.transform(hX_train)
hX_test = h_scaler.transform(hX_test)
ap_scaler = StandardScaler()
ap_scaler.fit(apX_train)
apX_train = ap_scaler.transform(apX_train)
apX_test = ap_scaler.transform(apX_test)
```

```python
h_model = RandomForestRegressor(random_state = 42, n_estimators = 300)
ap_model = RandomForestRegressor(random_state = 42, n_estimators = 300)
# h_model = GradientBoostingRegressor(random_state = 24, n_estimators = 200)
# ap_model = GradientBoostingRegressor(random_state = 24, n_estimators = 200)
h_model.fit(hX_train, hy_train)
ap_model.fit(apX_train, apy_train)
print("Model trained: OK.")
print(f"H. score: {round(h_model.score(hX_test, hy_test)*100,2)}")
print(f"Ap. score: {round(ap_model.score(apX_test, apy_test)*100,2)}")
```

```python
h_scaler = StandardScaler()
h_scaler.fit(hX_train)
hX_train = h_scaler.transform(hX_train)
hX_test = h_scaler.transform(hX_test)
ap_scaler = StandardScaler()
ap_scaler.fit(apX_train)
apX_train = ap_scaler.transform(apX_train)
apX_test = ap_scaler.transform(apX_test)
h_pipeline = Pipeline([
    ("poly", PolynomialFeatures(degree = 2)),
    ("model", LinearRegression())
])
ap_pipeline = Pipeline([
    ("poly", PolynomialFeatures(degree = 2)),
    ("model", LinearRegression())
])
h_pipeline.fit(hX_train, hy_train)
ap_pipeline.fit(apX_train, apy_train)
print("Model trained: OK.")
print(f"H. score: {h_pipeline.score(hX_test, hy_test)}")
print(f"Ap. score: {ap_pipeline.score(apX_test, apy_test)}")
```

```python
df = pd.read_csv("data/clean_dataset.csv")
df.type = df.type.apply(lambda x: 1 if x == "house" else 2)
df["subtype"] = df["subtype"].apply(lambda x: _subtype_replace(x))
```

```python
df
```

```python
train = df.dropna()
train.state = train.state.astype(int)
sy = train.state
sX = train.drop("state", axis = 1)
sX_pred = df[df.state.isna() == True].drop(["state"], axis = 1)
sX_train, sX_test, sy_train, sy_test = train_test_split(sX, sy, random_state = 24, test_size = 0.02)
classifier = RandomForestClassifier(random_state = 24, n_estimators = 200)
classifier.fit(sX_train, sy_train)
predictions = classifier.predict(sX_pred)
frame_append = pd.DataFrame(sX_pred)
frame_append["state"] = predictions
frame_append
```

```python
frame_append2 = df[df.state.isna() == False]
frame_append2
```

```python
frame_append2.state = frame_append2.state.astype(int)
```

```python
appended = pd.concat([frame_append, frame_append2])
appended
```

```python
train = appended.dropna()
train.state = train.state.astype(int)
sy = train.facades
sX = train.drop("facades", axis = 1)
sX_pred = appended[appended.facades.isna()].drop(["facades"], axis = 1)
sX_train, sX_test, sy_train, sy_test = train_test_split(sX, sy, random_state = 24, test_size = 0.2)
classifier = RandomForestClassifier(random_state = 24, n_estimators = 200)
classifier.fit(sX_train, sy_train)
f_preds = classifier.predict(sX_pred)
classifier.score(sX_test, sy_test)
```
