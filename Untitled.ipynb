```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
def _subtype_replace(str):
    match str:
        case "studio" | "chalet":
            return 1
        case "ground-floor" | "apartment" | "bungalow" | "triplex" | "duplex" | "residence":
            return 2
        case "mixed-building" | "loft" | "cottage" | "penthouse" | "master-house" | "villa":
            return 3
        case "mansion":
            return 4
```

```python
h_scaler = StandardScaler()
h_scaler.fit(hX_train)
hX_train = h_scaler.transform(hX_train)
hX_test = h_scaler.transform(hX_test)
ap_scaler = StandardScaler()
ap_scaler.fit(apX_train)
apX_train = ap_scaler.transform(apX_train)
apX_test = ap_scaler.transform(apX_test)
```

```python
def _subtype_fit(data, prices):
    df = pd.DataFrame(data).iloc[:,1].to_frame()
    df.columns = ["subtype"]
    df["price"] = prices
    subtypes = {}
    types_list = df.groupby("subtype")["price"].mean().sort_values().reset_index()["subtype"]
    for i, value in enumerate(types_list):
        subtypes[value] = i
    return subtypes

_subtypes = _subtype_fit(data, data.price)

def _subtype_transform(value):
    return _subtypes[value]
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("data/clean_dataset.csv")
df = df[["price", "living_area"]].sort_values(["price", "living_area"])
print(df.shape)
df.head()
```

```python
fig, ax = plt.subplots()
df = df.iloc[:-50,:]
print(df.shape)
ax.scatter(df.price, df.living_area)
plt.show()
```

scaler = StandardScaler()
df = scaler.fit_transform(df)
df

```python
kmeans = KMeans(n_clusters = 4, init = "k-means++", n_init="auto").fit(df)
```

```python
df = pd.DataFrame(df)
df.columns = ["price", "living_area"]
df["cluster"] = kmeans.labels_
df
```

```python
fig, ax = plt.subplots()
ax.scatter(df[df.cluster == 0].price, df[df.cluster == 0].living_area)
ax.scatter(df[df.cluster == 1].price, df[df.cluster == 1].living_area)
ax.scatter(df[df.cluster == 2].price, df[df.cluster == 2].living_area)
ax.scatter(df[df.cluster == 3].price, df[df.cluster == 3].living_area)
plt.show()
```
