```python
import pandas as pd
from geopy.distance import geodesic

LOCS = [
    [50.226094, 4.889102],
    [50.827060, 4.422676],
    [50.878292, 4.699861],
    [51.210784, 4.430718],
    [51.026489, 4.740961],
    [51.052728, 3.724531],
    [51.210671, 3.226417],
    [51.348528, 3.301504]
]
```

```python
cords = pd.read_csv("../data/external/cords.csv")
```

```python
cords["distance"] = 0.0
for i in range(len(cords)):
    short_find = []
    for j in range(len(LOCS)):
        short_find.append(round(geodesic((cords.x[i], cords.y[i]), (LOCS[j][0], LOCS[j][1])).km, 3))
    cords.at[i, "distance"] = min(short_find)
```
