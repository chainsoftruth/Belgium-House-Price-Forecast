```python
import pandas as pd
from geopy.distance import geodesic

LOCS = [
    [50.85, 4.349],
    [51.22, 4.4],
    [51.05, 3.717],
    [51.209, 3.224]
]
```

```python
cords = pd.read_csv("data/external/cords.csv")
```

```python
cords["distance"] = 0.0
for i in range(len(cords)):
    short_find = [
        round(geodesic((cords.x[i], cords.y[i]), (LOCS[0][0], LOCS[0][1])).km, 3),
        round(geodesic((cords.x[i], cords.y[i]), (LOCS[1][0], LOCS[1][1])).km, 3),
        round(geodesic((cords.x[i], cords.y[i]), (LOCS[2][0], LOCS[2][1])).km, 3),
        round(geodesic((cords.x[i], cords.y[i]), (LOCS[3][0], LOCS[3][1])).km, 3)
    ]
    cords.at[i, "distance"] = min(short_find)
```
