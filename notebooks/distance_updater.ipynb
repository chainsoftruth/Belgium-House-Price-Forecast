```python
import pandas as pd
from geopy.distance import geodesic

LOCS = [
    [50.831339, 4.403835],
    [51.22, 4.4],
    [51.05, 3.717],
    [51.209, 3.224],
    [50.6401, 5.5707],
    [51.354897, 3.284192],
    [51.285762, 3.050603],
    [51.212303, 2.887623],
    [51.160253, 2.765595],
    [51.119909, 2.643586],
    [50.669207, 4.568247]
]
```

```python
cords = pd.read_csv("../data/external/cords.csv")
```

```python
cords["distance"] = 0.0
for i in range(len(cords)):
    short_find = []
    for j in range(len(LOCS)):
        short_find.append(round(geodesic((cords.x[i], cords.y[i]), (LOCS[j][0], LOCS[j][1])).km, 3))
    cords.at[i, "distance"] = min(short_find)
```
